```
%%capture
import sys
sys.path.append("input-marginalization")
sys.path.append("input-marginalization/generative_model")
import math

!pip install transformers

import torch
import torch.nn as nn
import torch.nn.functional as F

import transformers
from transformers import BertForMaskedLM

from metrics import continuous_colored_sentence
from gen_model_metrics import input_marginalization
import pickle
```

```
model = torch.load(f"{save_dir}/transformer_wikitext2.pt")
```

```
%%capture
mlm = BertForMaskedLM.from_pretrained('bert-base-uncased').cuda()
```

```
sentences = [
  "sam i am",
  "less is more"
]
```

```
att_scores = []
for s, sentence in enumerate(sentences):
  print("Sentence:", s)
  att_scores.append(input_marginalization(model, sentence, mlm, num_batches = 20))

pickle.dump(att_scores, open(f"{save_dir}/gen_model_att_scores_fig2.p", "wb" ))
```

```
att_scores = pickle.load(open(f"{save_dir}/gen_model_att_scores_fig2.p", "rb" ))
for i, sentence in enumerate(sentences):
  continuous_colored_sentence(sentence, att_scores[i], color_limit=0.003)
  print("")
```
