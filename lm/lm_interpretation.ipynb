```
from transformers import GPT2Tokenizer
from torch.nn import LogSoftmax

log_softmax = LogSoftmax(dim=0)
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

def compute_log_prob(model, sentence, device="cuda"):
  tokens = torch.tensor(tokenizer(sentence).input_ids).to(device)
  with torch.no_grad():
      log_prob = -float(model(tokens, labels=tokens, return_dict=True).loss)

  return log_prob
```

```
sentence = "The cat drinks the milk."

print(compute_log_prob(gpt2, sentence))
```
